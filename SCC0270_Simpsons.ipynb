```python
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2 , preprocess_input
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
```

1.

```python
mconv = MobileNetV2(weights='imagenet', include_top=False, input_shape=(128, 128, 3))
```

```python
mconv.summary()
```

```python
for layer in mconv.layers:
    layer.trainable = False
```

```python
x = Flatten()(mconv.output)
output = Dense(7, activation='softmax')(x)

model = Model(inputs=mconv.input, outputs=output)
```

```python
model.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()
```

2.

```python
train_image = ImageDataGenerator(
  horizontal_flip=True,
 zoom_range=0.2,
 shear_range=10,
 height_shift_range=0.1,
  width_shift_range=0.1,
  preprocessing_function= preprocess_input
)
test_image = ImageDataGenerator(preprocessing_function=preprocess_input)
```

```python
train = train_image.flow_from_directory('/content/dataset/train', target_size=(128, 128), batch_size=32)
test = test_image.flow_from_directory('/content/dataset/test', target_size=(128, 128), batch_size=32,shuffle=False)
```

```python
resposta = model.fit(train, epochs=10, validation_data=test,batch_size=32,steps_per_epoch=70)
```

```python
resposta.history['accuracy'] # acuracia da base de train
```

```python
resposta.history['val_accuracy'] # acuracia da base de test
```
